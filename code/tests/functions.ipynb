```python
import math


def gaussian(x, mu=0, std=1):
    return 1 / (math.sqrt(math.pi * 2) * std) * math.exp(- 1/2 * (x - mu)**2 / std)

def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
import matplotlib
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
import numpy as np

n = 2000
k = 3

samples = np.random.rand(n) * (2 * k) - k  # -1 ~ 1
```

```python
ys = [10 * (sigmoid(gaussian(x, 0)) - 1/2) for x in samples]
```

```python
plt.scatter(samples, ys)
plt.show()
```

```python
def func(t=0, n=3000):
    std = 1
    mu = 0
    print(f"t={t}")

    samples = np.random.randn(n) * std + mu

    ys = [sigmoid(x) - t for x in samples]
    mask = [x >= 0 for x in ys]
    print(f"ratio={sum(mask) / len(mask)}")
    plt.hist(ys, bins=200, density=True)
    plt.show()
```

```python
func(0)
func(0.5)
func(0.6)
func(0.7)
```
